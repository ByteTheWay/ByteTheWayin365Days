# TextBlob

```python
from textblob import TextBlob

text = "ByteTheWay is a 365 days challenge to learn new codes in an \
        easy way. Today, we learn to count the frequency of a certain \
        word. TextBlob is a library to process textual data [Natural \
        Language Processing - NLP]. NLP is a subfield of computers \
        science and artificial intelligence that allows computer to \
        understand text and spoken human language as well as humans do."

blob = TextBlob(text)
print(blob.word_counts['to'])
```
