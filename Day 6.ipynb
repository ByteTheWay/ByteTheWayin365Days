```python
# Dictionary of percent body fat
d = {'Data 1': 12.3, "Data 2": 6.1, "Data 3": 25.3, "Data 4": 19.2}
print("Origin Dict: ", d)

d['Data 3'] = 11.7
print("Newer Dict: ", d)
```

```python
# List of percent body fat
l = [12.3, 6.1, 25.3, 19.2]
print("Origin List: ", l)

l[2] = 11.7
print("Newer List: ", l)
```

```python
# Tuple of percent body fat
t = (12.3, 6.1, 25.3, 19.2)
print("Origin Tuple: ", t)

t[2] = 11.7
print("Newer Tuple: ", t)
```
