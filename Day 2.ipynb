# Mutable and Immutable Object

```python
# Origin value
x = [2021]
print("Before: ")
print(hex(id(x)))
print(x)

# Changed value
x[0] = [2022]
print("\nAfter")
print(hex(id(x)))
print(x)
```

```python
#Origin value
x = 2021
print("Before: ")
print(hex(id(x)))
print(x)

#Changed value
x = 2022
print("\nAfter: ")
print(hex(id(x)))
print(x)
```

As you see in the figure shown below, a list is an object that has the same address even though its value is changed, and vice versa if that is int. The origin value of int 13 is no longer in use and that is a garbage that is automatically claimed by Garbage Collector from Python. FYI, every object is stored in the memory and they must be unique. So, waste can cause memory leaks and this can cause a serious problem for long-running applications. But, newer languages such as Python now have Automatic Memory Management to tackle this issue. No worries if you forgot to free your memory too late or too soon!
