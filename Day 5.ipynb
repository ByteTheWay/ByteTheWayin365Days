```python
from textblob import TextBlob

text = "Hello! i wil helep yoy tw write"

blob = TextBlob(text)
print(blob.correct())
```

TextBlob is a Python library to process textual information. For instance, in common NLP, we have a feature such as spelling correction. And then, TextBlob.correct() is a method to correct words that have been mistaken in a sentence. Let's give it a try!
