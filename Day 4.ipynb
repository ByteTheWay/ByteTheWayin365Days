```python
a = [1, 2, 3, 4, 5]
b = []

for x in a:
    if x%2==0:
        b.append("even")
    else:
        b.append("odd")

print(b)
```

```python
a = [1, 2, 3, 4, 5]

b = ['even' if i%2==0 else 'odd' for i in a]
print(b)
```

List Comprehension is a shorter syntax to make a new list based on an existing list. For instance, Code 1 is an ordinary syntax to categorise a number whether it is odd or even in a list. However, Code 2 has a shorter syntax to categorise them in accordance with:

newlist = [expression if condition == True for item in iterable ]

The expression in the Code 2 below says:
"Return even if it has no remainder, if it has remainder return odd"
